# 1. 당뇨병

```python
# 데이터 로드

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

diabetes = datasets.load_diabetes(as_frame=True)

df_X = diabetes.data
df_y = diabetes.target

print(type(diabetes))
print(type(df_X))
print(type(df_y))
```

```python
# 특성 선택을 위한 상관계수분석

origin = pd.concat([df_X, df_y], axis=1)
origin.corr(method='pearson')
```

```python
# 모델에 입력할 데이터 준비
# 특성 선택
features = ['bmi','bp','s3','s4','s5','s6']

X = df_X[features].values
y = df_y.values

print(type(X))
print(type(y))
```

```python
# 데이터 분할

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
# 가중치 설정 및 모델링

W = np.random.randn(len(features))
b = np.random.randn()

def model(X, W, b):
    # y_pred = W*X + b
    y_pred = b
    for i in range(X.shape[1]):
        y_pred += X[:, i] * W[i]
    return y_pred

print(W)
print(b)
```

```python
# 손실함수
def loss(X, W, b, y):
    y_pred = model(X, W, b)
    mse = ((y_pred - y) ** 2).mean()
    return mse
```

```python
# 기울기 함수
def gradient(X, W, b, y):
    # N은 데이터 포인트의 개수
    N = len(y)
    
    # y_pred 준비
    y_pred = model(X, W, b)
    
    # 공식에 맞게 gradient 계산
    dW = 1/N * 2 * X.T.dot(y_pred - y)
        
    # b의 gradient 계산
    db = 2 * (y_pred - y).mean()
    return dW, db
```

```python
# 학습률 지정 및 모델 학습
LEARNING_RATE = 0.1
losses = []

for i in range(1, 10001):
    dW, db = gradient(X_train, W, b, y_train)
    W -= LEARNING_RATE * dW
    b -= LEARNING_RATE * db
    L = loss(X_train, W, b, y_train)
    losses.append(L)
    if i % 1000 == 0:
        print('Iteration %d : Loss %0.4f' % (i, L))
```

```python
W, b
```

```python
# 모델 성능 평가
y_pred = model(X_test, W, b)
mse = loss(X_test, W, b, y_test)

mse
```

```python
# 시각화
plt.scatter(X_test[:, 0], y_test, color='orange')
plt.scatter(X_test[:, 0], y_pred, color='green')
plt.xlabel(features[0])
plt.ylabel('target')
# plt.legend()
plt.show()
```

# 2. 자전거

```python
# 데이터 로드
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

data = pd.read_csv('~/data/data/bike-sharing-demand/train.csv')
data.head()
```

```python
# 자료형 변환 및 컬럼 생성

train = data.copy()
train.info()
```

```python
train['datetime'] = pd.to_datetime(train['datetime'])
train.info()
```

```python
train['year'] = train['datetime'].dt.year
train['month'] = train['datetime'].dt.month
train['day'] = train['datetime'].dt.day
train['hour'] = train['datetime'].dt.hour
train['minute'] = train['datetime'].dt.minute
train['second'] = train['datetime'].dt.second

train.sample(5)
```

```python
# 데이터 개수 시각화
import seaborn as sns

plt.figure(figsize=(36,18))

plt.subplot(231)
sns.countplot(data=train, x='year').set_xlabel('')
plt.title('Year')

plt.subplot(232)
sns.countplot(data=train, x='month').set_xlabel('')
plt.title('Month')

plt.subplot(233)
sns.countplot(data=train, x='day').set_xlabel('')
plt.title('Day')

plt.subplot(234)
sns.countplot(data=train, x='hour').set_xlabel('')
plt.title('Hour')

plt.subplot(235)
sns.countplot(data=train, x='minute').set_xlabel('')
plt.title('Minute')

plt.subplot(236)
sns.countplot(data=train, x='second').set_xlabel('')
plt.title('Second')

plt.show()
```

```python
train.info()
```

```python
# 컬럼 선택 및 데이터 분할
# X = train[['season','holiday','workingday','weather','temp','atemp',
#            'humidity','windspeed','year','month','day','hour']].values
X = train[['season','holiday','workingday','weather','temp','atemp',
           'humidity','windspeed','month','day','hour']].values
y = train['count'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
# 모델 학습
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
# 예측값 출력 및 손실함수값 계산
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred)
rmse = mse**0.5

print(y_pred, f'\n => 예측값의 개수는 모두 {len(y_pred)}개 입니다.')
print('='*75)
print(mse, rmse)
```

```python
# 시각화
plt.figure(figsize=(20,10))

plt.subplot(121)
plt.scatter(X_test[:, 4], y_test, label="true", color='black')
plt.scatter(X_test[:, 4], y_pred, label="pred", color='pink')
plt.title('Temperature')
plt.legend()

plt.subplot(122)
plt.scatter(X_test[:, 6], y_test, label="true", color='royalblue')
plt.scatter(X_test[:, 6], y_pred, label="pred", color='yellow')
plt.title('Humidity')
plt.legend()

plt.show()
```
