```python
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import pandas as pd
import os

#그래프 데이터
csv_path = os.getenv("HOME") + "/aiffel/data_visualization/data/AMZN.csv"
data = pd.read_csv(csv_path, index_col=0, parse_dates=True)
price = data['Close']

#index_col=0: 이 매개변수는 데이터프레임의 행 인덱스로 사용할 열의 인덱스(0부터 시작)를 지정합니다. 여기서는 첫 번째 열을 행 인덱스로 사용하고 있습니다.
#parse_dates=True: 이 매개변수가 True로 설정되면, 날짜/시간 열을 파싱하여 datetime 형식으로 변환합니다.


# 축 그리기 및 좌표축 설정

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
price.plot(ax=ax, style = 'red')
plt.ylim([1600, 2200])
plt.xlim(['2019-05-01', '2020-03-01'])

#주석달기
#asof(d)는 시리즈나 데이터프레임에서 d에 가장 가까운 날짜나 시간의 값으로 쿼리하라는 의미입니다. 코드에서 price.asof(d)는 price라는 데이터프레임에서 
#d에 가장 가까운 시간의 값(가격)을 검색하라는 것을 의미합니다. 이 값을 기반으로 주석을 달 좌표를 조정하는 데 사용됩니다.
important_data = [(datetime(2019, 6, 3), "Low Price"), (datetime(2020, 2, 19), "Peak Price")]
for d, label in important_data:
    ax.annotate(label, xy=(d, price.asof(d)+10), # 주석을 달 좌표(x,y)
                xytext=(d, price.asof(d)+100), # 주석 텍스트가 위치할 좌표(x,y)
                arrowprops=dict(facecolor='red')) # 화살표 추가 및 색 설정
    
# 그리드, 타이틀 달기
plt.grid()
ax.set_title('StockPrice')

plt.show()


```
